```python
from imutils import face_utils
import numpy as np
import imutils
import dlib
import cv2
import keras
import winsound
```

```python
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import os
```

```python
frequency = 2500
duration = 200
```

```python
img_width=64
img_height=64
model = load_model('trained_model.h5')
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model
```

```python
out=list()
dif0=list()
dif1=list()
camera = cv2.VideoCapture(0)
predictor_path = 'shape_predictor_68_face_landmarks.dat'
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(predictor_path)
(lStart, lEnd) = face_utils.FACIAL_LANDMARKS_IDXS["left_eye"]
(rStart, rEnd) = face_utils.FACIAL_LANDMARKS_IDXS["right_eye"]
(mStart, mEnd) = face_utils.FACIAL_LANDMARKS_IDXS["mouth"]
count = 0
trail = 0
while True:
    ret, frame = camera.read(1)
    frame = imutils.resize(frame, width=640)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    rects = detector(gray, 0)
    for rect in rects:
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        leftEye = shape[lStart:lEnd]
        rightEye = shape[rStart:rEnd]
        mouth = shape[mStart:mEnd]
        (x, y, w, h) = cv2.boundingRect(np.array([shape[lStart:lEnd]]))
        roi = frame[y-10:y + h+10, x-10:x + w+10]
        cv2.imwrite("frame%d.jpg" % count, roi) 
        test_image=image.load_img("frame%d.jpg" % count,target_size=(64,64))
        test_image=image.img_to_array(test_image)
        test_image=np.expand_dims(test_image,axis=0)
        result=model.predict(test_image)
        os.remove("frame%d.jpg" % count)
        
        if result[0][0]<0.5:
            out.append(0)
            trail = trail + 1
            if trail>40:
                cv2.putText(frame, "Alert!!! Eyes Closed ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                winsound.Beep(frequency, duration)
                
        else :
            out.append(1)
            trail = 0
            cv2.putText(frame, "Eyes Open ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        count += 1
    
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
cv2.destroyAllWindows()
camera.release()

        
```
